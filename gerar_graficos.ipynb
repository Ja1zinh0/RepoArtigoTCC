```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np
```

```python
df_quest = pd.read_csv('questionario.csv')
```

```python
df_quest.head()
```

```python
# Visualizar nome das colunas
for collum in df_quest:
    print(collum)
```

```python
# Renomear colunas
df_quest = df_quest.rename(columns={
    '1. Qual a sua faixa etária?': 'Faixa Etaria',
    '2. Qual seu sexo?': 'Sexo',
    '3. Qual seu nível de escolaridade?': 'Escolaridade',
    '4. Selecione abaixo os seus gêneros literários preferidos:': 'Generos Literarios',
    '5. Quantos livros, em média, você lê por ano?': 'Livros por Ano',
    '6. Qual é a sua forma preferida de leitura?':'Forma de Leitura',
    '7. Em que tipo de ambiente você costuma ler?': 'Ambiente',
    '8. Qual é o seu maior obstáculo para ler um livro? (Dificuldades e limitações)': 'Dificuldade',
    '9. O que a leitura pode contribuir/contribuiu para você?': 'Contribuicao',
    '10. Qual foi o seu primeiro contato com a leitura?': 'Primeiro Contato',
    'PERGUNTA EXTRA\nO que você sugere para melhorar a experiência do leitor com os livros e bibliotecas?': 'Extra'
})
```

```python
df_quest.head()
```

```python
fig, axs = plt.subplots(1,1)

faixa_etaria_counts = df_quest['Faixa Etaria'].value_counts()

axs.pie(faixa_etaria_counts, autopct='%1.1f%%', startangle=90)
axs.legend(faixa_etaria_counts.index, title="Faixas Etárias", loc="upper right")

plt.show()
```

```python
fig, axs = plt.subplots(1,1)

sexo_counts = df_quest['Sexo'].value_counts()
axs.pie(sexo_counts, autopct='%1.1f%%', startangle=90)
axs.legend(sexo_counts.index, title="Sexos", loc="upper right")

plt.show()
```

```python
generos = df_quest['Generos Literarios'].str.split(';', expand=True).stack()

contagem = generos.value_counts()

total = contagem.sum()

plt.barh(contagem.index, contagem.values, height=0.5, color="DarkBlue")
porcentagens = (contagem / total) * 100

for i, (valor, percentual) in enumerate(zip(contagem.values, porcentagens.values)):
    plt.text(valor, i, f"{valor} ({percentual:.1f}%)", ha='left', va='center', fontsize=10)

plt.xlim(right=contagem.values.max() * 1.2)


plt.show()
```

```python
fig, axs = plt.subplots(1,1)

escolaridade_counts = df_quest['Escolaridade'].value_counts()
bars = axs.bar(escolaridade_counts.index, escolaridade_counts.values, color="DarkBlue")

total = escolaridade_counts.sum()
porcentagens = ((escolaridade_counts / total) * 100).round(1)


axs.tick_params(axis='x', labelrotation=45)
axs.set_xticks(range(len(escolaridade_counts.index)))
axs.set_xticklabels(escolaridade_counts.index, ha='right')

for bar, perc in zip(bars, porcentagens):
    height = bar.get_height()
    axs.annotate('{}%'.format(round(perc, 2)),
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 2),
                 textcoords="offset points",
                 ha='center', va='bottom')


plt.show()
```

```python
fig, axs = plt.subplots(1,1)

media_livros_counts = df_quest['Livros por Ano'].value_counts()

total = media_livros_counts.sum()
porcentagens = (media_livros_counts / total) * 100

bars = axs.bar(media_livros_counts.index, media_livros_counts.values, color="DarkBlue")

axs.tick_params(axis='x', labelrotation=45)
axs.set_xticks(range(len(media_livros_counts.index)))
axs.set_xticklabels(media_livros_counts.index, ha='right')

for bar, percentual in zip(bars, porcentagens):
    height = bar.get_height()
    axs.annotate('{:.1f}%'.format(percentual),
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 2),
                 textcoords="offset points",
                 ha='center', va='bottom')

plt.show()
```

```python
import re
dificuldades = df_quest['Dificuldade'].str.split(';', expand=True).stack()

dificuldades = dificuldades.str.lower()

dificuldades = dificuldades.apply(lambda x: 'tempo' if re.search(r'\btempo\b', x) else ('concentração' if re.search(r'\bconcentração\b', x) else ('dinheiro/preço' if re.search(r'\bdinheiro|preço\b', x) else x)))

contagem = dificuldades.value_counts()

contagem = contagem.head(3)

total = contagem.sum()

plt.barh(contagem.index, contagem.values, height=0.5, color="DarkBlue")

porcentagens = (contagem / total) * 100

for i, (valor, percentual) in enumerate(zip(contagem.values, porcentagens.values)):
    plt.text(valor, i, f"{valor} ({percentual:.1f}%)", ha='left', va='center', fontsize=10)

plt.xlim(right=contagem.values.max() * 1.2)

plt.show()
```
